```python
import cv2
import mediapipe as mp
from dollarpy import Recognizer, Template, Point
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
from dollarpy import Recognizer, Template, Point
```

```python
vid = ['straight.mp4','right.MP4','left.mp4']
classes=[[]]
for i  in range(3):
    cap = cv2.VideoCapture(vid[i])
    allpts=[]
    with mp_face_detection.FaceDetection(
        model_selection=0, min_detection_confidence=0.5) as face_detection:
        while cap.isOpened():
            success, image = cap.read()
            if not success:
                break

            image = cv2.flip(image, 1)

            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            results = face_detection.process(image_rgb)

            if results.detections:
                for detection in results.detections:
                    mp_drawing.draw_detection(image, detection)

                    keypoints = detection.location_data.relative_keypoints
                    for keypoint in keypoints:
                        x = int(keypoint.x * image.shape[1])
                        y = int(keypoint.y * image.shape[0])
                        p = Point(x,y,0)

                        allpts.append(p)
                        cv2.circle(image, (x, y), 5, (0, 255, 0), -1)
          

            cv2.imshow('Face Detection', image)

            if cv2.waitKey(5) &  0xFF == ord('q'):
                break
    print(allpts)
    classes.append(allpts)
# print(classes)
cap.release()
cv2.destroyAllWindows()
```

```python
arr=["Straight","Right","Left"]
temp=[]
for i in range(3):
    tmpl_1 = Template(arr[i],classes[i+1] )
    temp.append(tmpl_1)
recognizer = Recognizer(temp)


```

```python
def getpoints(path):



    cap = cv2.VideoCapture(path)
    allpts=[]
    with mp_face_detection.FaceDetection(
        model_selection=0, min_detection_confidence=0.5) as face_detection:
        while cap.isOpened():
            success, image = cap.read()
            if not success:
                break

            image = cv2.flip(image, 1)

            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            results = face_detection.process(image_rgb)

            if results.detections:
                for detection in results.detections:
                    mp_drawing.draw_detection(image, detection)

                    keypoints = detection.location_data.relative_keypoints
                    for keypoint in keypoints:
                        x = int(keypoint.x * image.shape[1])
                        y = int(keypoint.y * image.shape[0])
                        p  =Point(x,y,0)
                    # print(p)
                        allpts.append(p)
                        cv2.circle(image, (x, y), 5, (0, 255, 0), -1)
                        #cv2.imshow("Annotadet",image)

            cv2.imshow('Face Detection', image)

            if cv2.waitKey(5) &  0xFF == ord('q'):
                break
    print(allpts)

# print(classes)
    cap.release()
    cv2.destroyAllWindows()
    return allpts
```
